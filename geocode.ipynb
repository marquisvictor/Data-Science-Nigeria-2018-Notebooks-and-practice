### Loading Geographic data to csv

```python
geo = pd.read_csv('geocodemaps.csv')

#len(geo), len(custid)

#all_df = pd.concat([custid, geo], sort=False)

geo.head()
```

```python
import plotly.graph_objects as go

geo['text'] = geo['formatted_'].astype(str)

fig = go.Figure(data=go.Scattergeo(
        lon = geo['long'],
        lat = geo['lat'],
        text = geo['text'],
        mode = 'markers',
        marker_color = 'black',
        ))

fig.update_layout(
        title = 'Location of Access Bank Customers in Nigeria<br>(Hover on each dot for place names)',
        geo_scope = 'africa', autosize=False, width=1500, height=1500,
    )

fig.show()
```
